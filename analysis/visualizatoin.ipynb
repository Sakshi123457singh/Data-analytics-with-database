```python
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
import plotly.express as px
```

```python
penguins = sns.load_dataset("penguins")
penguins
```

```python
penguins['bill_length_mm'].plot()
```

```python
penguins['bill_length_mm'].plot(linestyle='',marker='o',markersize=5, alpha= 1, figsize=(15,7) , markerfacecolor='yellow',
                                markeredgecolor='black')
```

```python
penguins['bill_length_mm'][:50].plot(kind='bar',figsize=(15,2),color='green')
```

```python
penguins['bill_length_mm'][:50].plot(kind='area',figsize=(15,2),color='green',alpha=0.5)
```

```python
print(sns.get_dataset_names())
```

```python
penguins.plot(kind='scatter',x='bill_length_mm',y='bill_depth_mm',s=100,alpha=.5)
```

```python
penguins.plot(kind='scatter',x='bill_length_mm',y='flipper_length_mm',s=100,alpha=.5)
```

```python
#sbplots
fig, axs= plt.subplots(1,2,figsize=(15,5),constrained_layout=True)
penguins.plot(kind='scatter',x='bill_length_mm',y='flipper_length_mm',s=100,alpha=.5, ax=axs[0],title='penguins bill data')
penguins.plot(kind='scatter',x='bill_length_mm',y='flipper_length_mm',s=100,alpha=.5, ax=axs[1])
plt.show()#for showing only graph
```

```python
penguins['species'].value_counts().plot(kind='pie',autopct='%1.1f%%',# addpercentage
                                        figsize=(3,3),title='species',ylabel='', pctdistance=.7,wedgeprops=dict(
                                            width=0.5
                                        ))
```

Advanced visualizatoin   

```python
tips=sns.load_dataset('tips')
tips.head(2)
```

```python
sns.countplot(data=tips, x='sex')
```

```python
sns.countplot(data=tips,x='sex',hue='day')
```

```python
sns.displot(data=tips,x='total_bill',binwidth=5)#distribution plot (histogram)
```

```python
sns.displot(data=tips,x='total_bill', col='smoker', row='sex',aspect=2,height=2,kind='kde',hue='day',multiple='fill')
```

```python
sns.catplot(data=tips, x="day", y="total_bill", kind="box")
```

```python
diamonds = sns.load_dataset("diamonds")
sns.catplot(
    data=diamonds.sort_values("color"),
    x="color", y="price", kind="boxen",
)
```

```python
sns.displot(penguins, x="flipper_length_mm", hue="species", kind="kde", fill=True)
```

```python
sns.displot(penguins, x="bill_length_mm", y="bill_depth_mm", hue="species")
```

```python
g = sns.PairGrid(penguins)
g.map_upper(sns.histplot)
g.map_lower(sns.kdeplot, fill=True)
g.map_diag(sns.histplot, kde=True)
```

```python
tips = sns.load_dataset("tips")
sns.catplot(data=tips, x="day", y="total_bill")
```

```python
sns.catplot(data=tips, x="day", y="total_bill", hue="sex", kind="swarm")
```

```python
sns.catplot(
    data=tips, x="total_bill", y="day", hue="sex",
    kind="violin", bw=.15, cut=0,
)
```

```python
titanic = sns.load_dataset("titanic")
sns.catplot(data=titanic, x="sex", y="survived", hue="class", kind="bar")
```

```python
sns.catplot(data=titanic, x="deck", kind="count", palette="ch:.25")
```

```python
sns.catplot(
    data=titanic, x="class", y="survived", hue="sex",
    palette={"male": "g", "female": "m"},
    markers=["^", "o"], linestyles=["-", "--"],
    kind="point"
)
```

```python
sns.catplot(
    data=tips, x="day", y="total_bill", hue="smoker",
    kind="swarm", col="time", aspect=.7,
)
```

```python
g = sns.catplot(
    data=titanic,
    x="fare", y="embark_town", row="class",
    kind="box", orient="h",
    sharex=False, margin_titles=True,
    height=1.5, aspect=4,
)
g.set(xlabel="Fare", ylabel="")
g.set_titles(row_template="{row_name} class")
for ax in g.axes.flat:
    ax.xaxis.set_major_formatter('${x:.0f}')
```
