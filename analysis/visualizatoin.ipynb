```python
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
import plotly.express as px
```

```python
penguins = sns.load_dataset("penguins")
penguins
```

```python
penguins['bill_length_mm'].plot()
```

```python
penguins['bill_length_mm'].plot(linestyle='',marker='o',markersize=5, alpha= 1, figsize=(15,7) , markerfacecolor='yellow',
                                markeredgecolor='black')
```

```python
penguins['bill_length_mm'][:50].plot(kind='bar',figsize=(15,2),color='green')
```

```python
penguins['bill_length_mm'][:50].plot(kind='area',figsize=(15,2),color='green',alpha=0.5)
```

```python
print(sns.get_dataset_names())
```

```python
penguins.plot(kind='scatter',x='bill_length_mm',y='bill_depth_mm',s=100,alpha=.5)
```

```python
penguins.plot(kind='scatter',x='bill_length_mm',y='flipper_length_mm',s=100,alpha=.5)
```

```python
#sbplots
fig, axs= plt.subplots(1,2,figsize=(15,5),constrained_layout=True)
penguins.plot(kind='scatter',x='bill_length_mm',y='flipper_length_mm',s=100,alpha=.5, ax=axs[0],title='penguins bill data')
penguins.plot(kind='scatter',x='bill_length_mm',y='flipper_length_mm',s=100,alpha=.5, ax=axs[1])
plt.show()#for showing only graph
```

```python
penguins['species'].value_counts().plot(kind='pie',autopct='%1.1f%%',# addpercentage
                                        figsize=(3,3),title='species',ylabel='', pctdistance=.7,wedgeprops=dict(
                                            width=0.5
                                        ))
```
